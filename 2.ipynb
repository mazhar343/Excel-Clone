```python
url='https://raw.githubusercontent.com/AmenaNajeeb/Data/master/data.csv'
```

```python
Fare_by_Age= df.groupby('Age').Fare.mean()[9:10]

Fare_by_Age
```

```python
x= Fare_by_Age.index.tolist()
y= Fare_by_Age.values.tolist()

plt.plot(x,y)
```

```python

print(y)
```

```python
Fare_by_Survived=df.groupby('Survived').Fare.mean()
Fare_by_Survived
```

```python
x= Fare_by_Survived.index.tolist()
y=Fare_by_Survived.values.tolist()

plt.bar(x,y)
```

```python
import numpy as np
```

```python
x= Fare_by_Survived.index.tolist()
y=Fare_by_Survived.values.tolist()

plt.bar(x,y)
plt.xticks(label=x,ticks=np.arange(len(x)))
plt.xlabel('person survived')
plt.ylabel('Fare')
plt.title('Fare by Survived')
```
